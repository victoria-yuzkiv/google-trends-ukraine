# Import necessary libraries

```python
import pandas as pd
import glob
import altair as alt
```

# Open and merge data

```python
# Get all files in the data directory

files = glob.glob("data/*.csv")
print(files)
```

```python
# Create a datafame and add the column that's gonna be a common one while merging data 

final = pd.DataFrame() 

data = pd.read_csv(files[0])
data, data.columns = data[1:] , data.iloc[0]
final['Тиждень'] = data.index
```

```python
# Define a function that opens new data and adds it to the final dataframe

def open_and_merge(new_filename, df_final):
    df = pd.read_csv(new_filename)
    df, df.columns = df[1:] , df.iloc[0]
    df['Тиждень'] = df.index
    return df_final.merge(df, on='Тиждень')
```

```python
# Process all files in order to get the final dataframe

for file in files:
    final = open_and_merge(file, final)
```
